```python
actions = []

for x in range(2):
    for y in range(2):
        for z in range(5):
            actions.append((x,y,z))
states = [i for i in range(6)]
rewards_table = np.zeros(shape=(len(actions),len(states)))
episode = 100000
loop = 10000


def determine_action(observation,rewards_table):
    
    best_action_index = np.argmax(rewards_table[:,observation])
    random_index = np.random.choice(states)    
    final_index = np.random.choice([best_action_index,random_index],p=[0.9,0.1]) #Explore randomly, but sometime exploit the best action 
    actions[final_index]

    return actions[final_index]

for e in range(episode):
    observation = env.reset()
    done = False
    for t in range(loop):

        action = determine_action(observation,rewards_table)
        new_observation, reward, done, info = env.step(action)
        #print(observation)
        action_index = actions.index(action)
        observation_index = states.index(observation)
        
        rewards_table[action_index][observation_index] += reward
        observation = new_observation
        #print(rewards_table)
        #env.render()
        if done:
            break

print(rewards_table)
```
